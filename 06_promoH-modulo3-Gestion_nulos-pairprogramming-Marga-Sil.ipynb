```python

import re
# importamos las librerías que necesitamos

# Tratamiento de datos
# -----------------------------------------------------------------------
import pandas as pd
import numpy as np

# Imputación de nulos usando métodos avanzados estadísticos
# -----------------------------------------------------------------------
from sklearn.impute import SimpleImputer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.impute import KNNImputer

# Librerías de visualización
# -----------------------------------------------------------------------
import seaborn as sns
import matplotlib.pyplot as plt
# Configuración
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames
```

```python
#importamos las tablas 
url='wdata_final.csv'
df_wd_part_final=pd.read_csv(url, usecols=lambda column: column != 'Unnamed: 0')
pd.set_option('display.max_columns', None)
```

```python
df_wd_part_final.head(2)
```

```python
#1. Identificación de Valores Nulos: Identifica todas las columnas que contengan valores nulos en el DataFrame.
df_wd_part_final_nulos=(df_wd_part_final.isnull().sum()/df_wd_part_final.shape[0])*100
df_wd_part_final_nulos
```

```python
#2. Selección de Método de Imputación: Discute en tu compañera cuál sería la mejor estrategia para manejar los valores nulos en cada una de las columnas identificadas en el paso anterior.

df_wd_part_final.info()
```

```python
#separamos variables categoricas y variabbles numericas:
df_wd_part_final_numericas= df_wd_part_final[df_wd_part_final.columns[df_wd_part_final.isnull().any()]].select_dtypes(include='float64')
df_wd_part_final_numericas
```

Metodos de imputacion variables numericas:
- si tiene mas de un 5% de nulos usamos IterativeImputer o KNNImputer
- si tiene menos de un 5% nulos usamos simple Imputer


```python
df_wd_part_final_nulos_num=(df_wd_part_final_numericas.isnull().sum()/df_wd_part_final_numericas.shape[0])*100
df_wd_part_final_nulos_num
```

```python
df_wd_part_final_nulos_num_filtro_menores=df_wd_part_final_nulos_num[df_wd_part_final_nulos_num < 5]
-df_wd_part_final_nulos_num_filtro_menores
```

```python
columnas_num_menor=list(df_wd_part_final_nulos_num_filtro_menores.keys())
```

```python
df_wd_part_final_num_menor=df_wd_part_final[columnas_num_menor]
df_wd_part_final_num_menor
```

```python
df_wd_part_final_num_menor.describe().T
```

```python
#usamos mediana para las siguientes columnas:
lista_mediana=['calling_code','infant_mortality']
for column in lista_mediana:
    imputer=SimpleImputer(strategy='median')
    column_imputer=imputer.fit_transform(df_wd_part_final[[column]]) #importante poner corchetes dobles!!
    df_wd_part_final[column]=column_imputer
```
